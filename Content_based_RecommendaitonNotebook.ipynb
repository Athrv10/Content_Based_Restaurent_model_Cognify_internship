```python
# fro that we need to apply the TF*IDF algorithm for the NNP and converting the documents into the vector format and
#document matrix
# also we need to create the sparce matrix for the scorecount .
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfvec = TfidfVectorizer(min_df=3,max_features=None,
                           token_pattern=r'\w{1}',ngram_range=(1,3),stop_words='english')
df['Cuisines'] = df['Cuisines'].fillna(" ")

tfidf_matrix = tfidfvec.fit_transform(df['Cuisines'])
```

```python
from sklearn.metrics.pairwise import linear_kernel,sigmoid_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
sig = sigmoid_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_rec(title, sig=sig):
    #getting the index of the title of restaurent
    idx = indices[title]

    # getting the pairwise similarity of the title of the restaurent
    sig_score = list(enumerate(sig[idx]))

    # soting the score
    sig_score = sorted(sig_score, key=lambda x:x[1],reverse=True)

    # getting the top 10 similar restaurents
    sig_score = sig_score[0:10]

    restaurents =[i[0] for i in sig_score]

    return df[['Restaurant Name', 'Cuisines', 'Aggregate rating']].iloc[restaurents]
```

```python
def recommend_by_user_preferences(preferred_cuisines, top_n=10):

    # Convert user preferences to a single string
    user_pref_string = ' '.join(preferred_cuisines).lower()

    # Transform user preferences into a vector
    user_pref_vector = tfidfvec.transform([user_pref_string])

    # Compute similarity between user preferences and all restaurants
    similarity_scores = sigmoid_kernel(user_pref_vector, tfidf_matrix).flatten()

    # Get top N indices based on similarity scores
    top_indices = similarity_scores.argsort()[::-1][:top_n]

    # Return restaurant details
    return df[['Restaurant Name', 'Cuisines', 'Aggregate rating']].iloc[top_indices]
```

```python
recommend_by_user_preferences("Thai,Japanese,Korean").sort_values("Aggregate rating", ascending=False)
```
